### 下載檔案

```python
import urllib.request
res = "https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-D0047-093?Authorization=CWB-73638260-F88A-444D-837B-498A220E9B8E&downloadType=WEB&format=ZIP"
urllib.request.urlretrieve(res, './data/F-D0047-093.zip')
```

### 解壓縮

```python
import zipfile
f = zipfile.ZipFile('./data/F-D0047-093.zip')
f.extractall('./data')
```

### 開檔/讀檔

```python
fh = open("./data/63_Week24_CH.xml", "r")
xml = fh.read()
fh.close()
```

```python
import os
data = []
path = './data/'
file = os.listdir(path)
file.remove('TAIWAN_Week24_CH.xml')
for k in file:
    if '_Week24_CH' in k:
        data.append(k)
```

```python
from bs4 import BeautifulSoup
import xmltodict
location = {}
city = {}
result = []
for l in data:
    fh = open(path+l, "r")
    xml = fh.read()
    d = dict(xmltodict.parse(xml))
    soup = BeautifulSoup(xml,'xml')
    town = []
    for j in range(0,len(soup.find_all('locationName'))):        
        week_temp = []
        for i in range(0,6):
            #抓到溫度
            temp = d['cwbopendata']['dataset']['locations']['location'][j]['weatherElement'][0]['time'][i]['elementValue']['value']
            #存成一週list
            week_temp.append(temp)
        location = {d['cwbopendata']['dataset']['locations']['location'][j]['locationName'] :week_temp} 
        town.append(location)
        city = {d['cwbopendata']['dataset']['locations']['locationsName']:town}
    result.append(city)
```
